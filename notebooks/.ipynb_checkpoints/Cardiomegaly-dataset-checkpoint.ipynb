## Cardiomegaly Overview

**Cardiomegaly** refers to an **enlargement of the heart**, which might be to heart failures

### Causes
1. **High Blood Pressure**
2. **Heart Valve Disease**
3. **Cardiomyopathy**

### Symptoms
- Shortness of breath
- Swelling in the legs
- Fatigue

### Diagnosis
- **Chest X-ray**
- **Echocardiogram**

```python
def read_data(data_set):
    cardiomegaly_images=[]
    cardiomegaly_labels=[]
    
    train_images=[]
    train_labels=[]
    
    test_images=[]
    test_labels=[]
    
    train_true_images=[]
    train_true_labels=[]

    train_false_images=[]
    train_false_labels=[]

    test_true_images=[]
    test_true_labels=[]

    test_false_images=[]
    test_false_labels=[]
    
    
    directories=os.listdir(data_set)
    for dir in directories:
        directory=os.path.join(Cardiomegaly_data_Set,dir)
        
        sub_directories=os.listdir(directory)
        for sub_dir in sub_directories:
           sub_directory=os.path.join(directory,sub_dir)
           images=glob.glob(os.path.join(sub_directory,'*.png'))
           labels=len(images) *[sub_dir]

           if(dir=='train' and sub_dir=='true'):
               train_true_images.extend(images)
               train_true_labels.extend(labels)

           elif(dir=='train' and sub_dir=='false'):
               train_false_images.extend(images)
               train_false_labels.extend(labels)

           elif(dir=='test' and sub_dir=='true'):
               test_true_images.extend(images)
               test_true_labels.extend(labels)

           elif(dir=='test' and sub_dir=='false'):
               test_false_images.extend(images)
               test_false_labels.extend(labels)

        if(dir=='train'):
            train_images.extend(train_true_images)
            train_images.extend(train_false_images)
            train_labels.extend(train_true_labels)
            train_labels.extend(train_false_labels)

        if(dir=='test'):
            test_images.extend(test_true_images)
            test_images.extend(test_false_images)
            test_labels.extend(test_true_labels)
            test_labels.extend(test_false_labels)
            
        
    cardiomegaly_images.extend(train_images)
    cardiomegaly_images.extend(test_images)
    
    cardiomegaly_labels.extend(train_labels)
    cardiomegaly_labels.extend(test_labels)
    
    return cardiomegaly_images,cardiomegaly_labels,train_images,train_labels,test_images,test_labels,train_true_images,train_true_labels,train_false_images,train_false_labels,test_true_images,test_true_labels,test_false_images,test_false_labels

        
```

```python
def get_distinct_image_size(data_set):
    img_sizes=[]
    
    for image in data_set:
        img=Image.open(image)
        size=img.size

        if(size in img_sizes):
            continue

        else:
           img_sizes.append(size)

    return img_sizes
        
```

# II. Data Preprocessing

```python
import cv2
import numpy as np
```

```python

def image_preprocess(images,crop_height=96,crop_width=96):

    image_list=[]
    for image in images:
        img=cv2.imread(image,cv2.IMREAD_GRAYSCALE)
        img=np.expand_dims(img,axis=-1)
        image_list.append(img)
    return np.array(image_list,dtype=np.float32) / 255
```

```python
def data_preprocess(train_images,train_labels,test_images,test_labels):
    
    train_labels=np.array(train_labels)
    test_labels=np.array(test_labels)
    X_train=image_preprocess(train_images)
    train_labels=train_labels=='true'
    y_train=train_labels.astype(np.float32)

    X_test=image_preprocess(test_images)
    test_labels=test_labels=='true'
    y_test=test_labels.astype(np.float32)

    return X_train,y_train,X_test,y_test

```

```python
def load_data():
    X_train,y_train,X_test,y_test=data_preprocess(train_images,train_labels,test_images,test_labels)
    return X_train,y_train,X_test,y_test
```

# III. Craete Model CNN

```python
from tensorflow.keras import Sequential
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model
from tensorflow.keras import backend as K
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import BatchNormalization
```

#### This model consists of:
 *  4 Convolutional Layers follwed by 4 Pooling Layers
 *  Hidden Fully Connected Layer(Dense Layer)
 *  2 Dropout Layers
 *  Output Layer

#### Relu function is used for hidden layers

  * ##### f(x)=max(0,x)
  
     where
      f(x)={
             
             if x>0 ,x
  
             if x<=0,0
            } 
            
   * ##### Range(0,infinity)

#### Sigmoid function  is used for output layer
   * ##### since range is (0,1) good for  binary classification


```python
def create_CNN_model(optimizer):
    model=Sequential(name='cardiomegaly-CNN-model')
   
    model.add(layers.Conv2D(32,(3,3),activation='relu',padding='same',input_shape=(128,128,1)))
    model.add(layers.MaxPooling2D(2,2))

    model.add(layers.Conv2D(64,(3,3),activation='relu',padding='same'))
    model.add(layers.MaxPooling2D(2,2))

    model.add(layers.Conv2D(128,(3,3),activation='relu',padding='same'))
    model.add(layers.MaxPooling2D(2,2))

    model.add(layers.Conv2D(128,(3,3),activation='relu',padding='same'))
    model.add(layers.MaxPooling2D(2,2))

    model.add(layers.Dropout(0.3))
    
    model.add(layers.Flatten())

    model.add(layers.Dense(128,activation='relu',kernel_regularizer=l2(0.01)))
    model.add(layers.Dropout(0.4))
    model.add(layers.Dense(1,activation='sigmoid'))

    model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return model
              
              
    
```

```python
from tensorflow.keras.optimizers import Adam
optimizer = Adam(learning_rate=0.001)
model=create_CNN_model(optimizer)
```

```python
from rich.table import Table
from rich.console import Console

def print_summary(model,model_name):
    console=Console()
    table=Table(title=model_name)

    table.add_column("Layer",style='green',justify='left')
    table.add_column("Input Shape",style='green',justify='left')
    table.add_column("Output Shape",style='green',justify='left')
    table.add_column("No of Parameters",style='green',justify='left')

    for layer in model.layers:
        layer_name=layer.name
        input_shape=str(layer.input_shape)
        output_shape=str(layer.output_shape)
        no_of_parameters=str(layer.count_params())
        table.add_row(layer_name,input_shape,output_shape,no_of_parameters)

    console.print(table)
```

```python
def calculate_measures(model):
    
    # get the loss and accuracy
    loss,accuracy=model.evaluate(X_test,y_test)

    # get the predicted y test
    y_pred_prob=model.predict(X_test)
    y_pred = (y_pred_prob > 0.5).astype(int)
    
    precision=precision_score(y_test,y_pred)
    recall= recall_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)
    print(f'Test Accuracy :{accuracy* 100:.2f}%') 
    console = Console()
    table = Table(title='Model')
    
    table.add_column("Model",style="black",justify="left")
    table.add_column("Accuracy",style="cyan",justify="left")
    table.add_column("Loss",style="magenta",justify="left")
    table.add_column("Precision",style="magenta",justify="left")
    table.add_column("Recall",style="green",justify="left")
    table.add_column("F1-Score",style="green",justify="left")
    table.add_row(model.name,str(accuracy),str(loss),str(precision),str(recall),str(f1))
    console.print(table)
    return y_test,y_pred
    
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
def plot_confusion_metrix(y_test,y_pred):
     cm=confusion_matrix(y_test,y_pred)
     plt.figure(figsize=(4,3))
     sns.heatmap(cm,cmap='Blues',annot=True,fmt='d',xticklabels=['Non-Cardiomegaly','Cardiomegaly'],yticklabels=['Non-Cardiomegaly','Cardiomegaly'])
     plt.tight_layout()
     plt.show()
    
```

```python
def plt_accuracy_and_loss(history):
    
    train_accuracy=history.history['accuracy']
    train_loss=history.history['loss']

    validation_accuracy=history.history['val_accuracy']
    validation_loss=history.history['val_loss']

    miminum_loss_index=np.argmin(validation_loss)
    minimum_loss=validation_loss[miminum_loss_index]

    maximum_accuracy_index=np.argmax(validation_accuracy)
    maximum_accuracy=validation_accuracy[maximum_accuracy_index]

    Epochs=[]
    for index in range(len(train_accuracy)):
        epoch=index+1
        Epochs.append(epoch)
 
    loss_label=f'Best Epochs : {str(miminum_loss_index)}'
    acc_label=f'Best Epochs : {str(maximum_accuracy_index)}'

    
    figure,axes=plt.subplots(1,2,figsize=(10,4))
    # plot train and validation accuracy
    axes[0].plot(Epochs,train_accuracy,'g',label='Train_accuracy')
    axes[0].plot(Epochs,validation_accuracy,'r',label='Validation_accuracy')

    axes[0].scatter(maximum_accuracy_index + 1 , maximum_accuracy , s = 100 , c = 'blue',label = acc_label)
    axes[0].set_title('Train and validation accuracy')

    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()

    # plot train and validation loss
    axes[1].plot(Epochs,train_loss,'g',label='Train_loss')
    axes[1].plot(Epochs,validation_loss,'r',label='Validation_loss')

    axes[1].scatter(miminum_loss_index + 1 , minimum_loss , s = 100 , c = 'blue',label = loss_label)
    axes[1].set_title('Train and validation loss')

    axes[1].set_xlabel('Epochs')
    axes[1].set_ylabel('Loss')
    axes[1].legend()

    plt.tight_layout()
    plt.show()
    
    
        
      

    
```

```python
def model_evaluation(model,history):
    plt_accuracy_and_loss(history)
    y_test,y_pred=calculate_measures(model)
    plot_confusion_metrix(y_test,y_pred)
    
    
```

### 3. Train and evaluation with cross fold validation

```python
from sklearn.model_selection import KFold
from tensorflow.keras.utils import to_categorical
from matplotlib.cm import get_cmap
from sklearn.metrics import ConfusionMatrixDisplay
K.clear_session()
```

#### Train

```python
import numpy as np

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.callbacks import EarlyStopping


# Define K-fold cross-validation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Lists to store performance metrics across folds
fold_accuracies = []
fold_precisions = []
fold_recalls = []
fold_f1_scores = []
fold_losses = []
train_accuracies=[]
val_accuracies=[]
train_losses=[]
val_losses=[]
cross_fold_models=[]
cross_fold_histories=[]
models=[]


X_train,y_train,X_test,y_test=load_data()
# K-Fold Cross Validation
for fold, (train_index, val_index) in enumerate(kf.split(X_train), start=1):
    print(f"\nStarting Fold {fold}...")
    
    # Split the data into training and validation sets for this fold
    X_train_data, X_val_data = X_train[train_index], X_train[val_index]
    y_train_data, y_val_data = y_train[train_index], y_train[val_index]

    optimizer = Adam(learning_rate=0.0001)
    # Create a new model instance for each fold
    model_cross_fold = create_CNN_model(optimizer)
    cross_fold_models.append(model_cross_fold)
    
    # Train the model
    history_cross_fold = model_cross_fold.fit(X_train_data, y_train_data, epochs=100,batch_size=128,validation_data=(X_val_data, y_val_data),callbacks=[early_stopping])
    cross_fold_histories.append(history_cross_fold)

    train_accuracies.append(history_cross_fold.history['accuracy'])
    val_accuracies.append(history_cross_fold.history['val_accuracy'])
    train_losses.append(history_cross_fold.history['loss'])
    val_losses.append(history_cross_fold.history['val_loss'])

    
    # Get predictions and compute additional metrics
    y_val_pred = (model_cross_fold.predict(X_val_data) > 0.5).astype(int)  # Binarize predictions if binary classification

    # Calculate metrics
    loss,accuracy=model_cross_fold.evaluate(X_test,y_test)

    # get the predicted y test
    y_pred_prob=model_cross_fold.predict(X_test)
    y_pred = (y_pred_prob > 0.5).astype(int)
    
    precision=precision_score(y_test,y_pred)
    recall= recall_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)
    print(f'Test Accuracy :{accuracy* 100:.2f}%') 
    console = Console()
    table = Table(title='Model')
    
    table.add_column("Accuracy",style="cyan",justify="left")
    table.add_column("Loss",style="magenta",justify="left")
    table.add_column("Precision",style="magenta",justify="left")
    table.add_column("Recall",style="green",justify="left")
    table.add_column("F1-Score",style="green",justify="left")
    table.add_row(str(accuracy),str(loss),str(precision),str(recall),str(f1))
    console.print(table)

    models.append(model_cross_fold)

    # Store metrics for each fold
    fold_precisions.append(precision)
    fold_recalls.append(recall)
    fold_f1_scores.append(f1)
    fold_accuracies.append(accuracy)
    fold_losses.append(loss)
    


# Calculate and print average metrics across all folds
avg_accuracy = np.mean(fold_accuracies)
avg_loss = np.mean(fold_losses)
avg_precision = np.mean(fold_precisions)
avg_recall = np.mean(fold_recalls)
avg_f1_score = np.mean(fold_f1_scores)

```

```python

# Plot training accuracy for each fold

def plot_folds_accuracies():
    
    cmap = get_cmap('tab10')
    numOfFolds=5
    numOfCols=3
    fig, axs = plt.subplots(2, numOfCols, figsize=(20,10))
    fig.suptitle('Training and Validation Accuracy per Fold')
    
    for i in range(numOfFolds):
        row=i//numOfCols
        if(i>2):
          col=i-(numOfCols)
        else:
          col=i
        ax=axs[row,col]
        
        color = cmap(i)  # Unique color for each fold
        Epochs=[]
        for index in range(len(train_accuracies[i])):
             epoch=index+1
             Epochs.append(epoch)
            
        ax.plot(Epochs, train_accuracies[i], label=f'Train Fold {i+1}', color=color, linestyle='-', linewidth=1.5)
    
        # Plot validation accuracy with dashed line
        ax.plot(Epochs, val_accuracies[i], label=f'Validation Fold {i+1}', color=color, linestyle='--', linewidth=1.5)
    
        ax.set_title(f'Fold {i+1}')  
        ax.set_xlabel('Epochs')  
        ax.set_ylabel('Accuracy')  
        ax.legend(loc='lower right')
        ax.grid(True)
    
    
    plt.subplots_adjust(hspace=0.5)
    axs[1,2].set_visible(False)
    plt.show()
```

```python
# Plot training accuracy for each fold
def plot_folds_losses():

    cmap = get_cmap('tab10')
    numOfFolds=5
    numOfCols=3
    fig, axs = plt.subplots(2, numOfCols, figsize=(20,10))
    fig.suptitle('Training and Validation Loss per Fold')
    
    for i in range(numOfFolds):
        row=i//numOfCols
        if(i>2):
          col=i-(numOfCols)
        else:
          col=i
        ax=axs[row,col]
        
        color = cmap(i)  # Unique color for each fold
        
        Epochs=[]
        for index in range(len(train_accuracies[i])):
             epoch=index+1
             Epochs.append(epoch)

        # Plot validation loss with solid line
        ax.plot(Epochs, train_losses[i], label=f'Train Fold {i+1}', color=color, linestyle='-', linewidth=1.5)
    
        # Plot validation loss with dashed line
        ax.plot(Epochs, val_losses[i], label=f'Validation Fold {i+1}', color=color, linestyle='--', linewidth=1.5)
    
        ax.set_title(f'Fold {i+1}')  # Title for each subplot
        ax.set_xlabel('Epochs')      # Label for x-axis
        ax.set_ylabel('Loss')    # Label for y-axis
        ax.legend(loc='upper right') # Legend to show which line corresponds to which
        ax.grid(True)  # Grid for better readability
    
    
    plt.subplots_adjust(hspace=0.5)
    axs[1,2].set_visible(False)
    plt.show()
```

```python

# Plot training accuracy for each fold

def plot_confusion_metrix():
    numOfFolds=5
    numOfCols=3
    fig, axs = plt.subplots(2, numOfCols, figsize=(20,20))
    fig.suptitle('Confusion matrix')
    
    for i in range(numOfFolds):
        row=i//numOfCols
        if(i>2):
          col=i-(numOfCols)
        else:
          col=i
        ax=axs[row,col]
         # Unique color for each fold

        X_train,y_train,X_test,y_test=load_data()
        y_pred_prob = cross_fold_models[i].predict(X_test)
        
        # Convert probabilities to binary predictions (0 or 1)
        y_pred = (y_pred_prob > 0.5).astype(int)
        
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["True", "False"])
        disp.plot(ax=ax, cmap=plt.cm.Blues)
       
        ax.set_title(f'Fold {i+1}')  # Title for each subplot
        ax.grid(True)  # Grid for better readability
    
    
    plt.subplots_adjust(hspace=0.2)
    axs[1,2].set_visible(False)
    plt.show()
```

```python
plot_folds_accuracies()
plot_folds_losses()
plot_confusion_metrix()
```
